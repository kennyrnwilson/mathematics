# Central Tendency
This sheet covers some basic statistics. We use the five year history of goofle

```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np

# First read in the 1 year price history for Google and store in a DataFrame
df = pd.read_csv("./data/GOOG.csv", index_col="Date")
```

## Measures
Measures of central tendency provide a single number that describes a whole data set
### Mean
The mean is the classic average value.

$$ \mu = \frac{\sum_{i=1}^n{x_i}}{n} $$ 

For our google prices we get

```python
print(f"Mean {df['Close'].mean()}")
```

### Median
The median is the middle value when we arrange the data in ascending or descending order. 

**Note:** We assume that we index the entries in the dataset such that the first element has index 1 (Unlike most programming languages where the first element has index 0)

 $$ median = \begin{cases} X[\frac{n+1}{2}] & \text{if n is odd} \\
 \frac{X[\frac{n+1}{2}]+X[\frac{n-1}{2}]}{2} & \text{if n is even} \end{cases} $$
 
 Also note that the Python Pandas Series object takes care of the sorting for us. 

## Symmetry
Conside the following data points which form a symmetrical distribution

```python
data = pd.Series((4, 5, 6, 6, 6, 7, 7, 7, 7, 7, 7, 8, 8, 8, 9, 10))
plt.hist(data,bins=7);
```

For our microsoft data we get

```python
print(f"Median {df['Close'].median()}")
```

```python
data = pd.Series((1,2,2,3))
plt.hist(data);
print(f"Median {data.median()}")
print(f"Mean {data.mean()}")
```

```python
data = pd.Series((1,2,2,3,10))
plt.hist(data);
print(f"Median {data.median()}")
print(f"Mean {data.mean()}")
```
